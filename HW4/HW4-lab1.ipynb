```python
import numpy as np
from scipy.misc import derivative
from sklearn.metrics import mean_squared_error
```

```python

X = [(1, -2, -1,-1,1), (2, -1,2,0, -2), (-1, 0,2,2, 1)]
Y = [7, 1, 1]
X = np.array(X)
Y = np.array(Y).reshape(-1,1)

N=X.shape[0]
dim=X.shape[1]
```

```python
def get_error(W,X,D):
    return D-np.dot(X,W)
def Weight_update(lr,W_old,X,D):
    error=get_error(W_old,X,D)
    W_new = W_old+lr*2*np.dot(X.T,error)
    return W_new
def get_Loss(D,W,X):
    y = np.dot(X,W)
    square=np.power(D-y, 2)

    return np.sum(square)
```

```python
epoch = 200
lr = 0.02
Loss_epoch =np.zeros([epoch ,1])
print(Loss_epoch.shape)
weight_epoch = np.zeros([epoch,dim])
print(weight_epoch.shape)
W_old=np.array([0,0,0,0,0]).reshape(dim,1)
```

```python
def gra_de(W_old,lr,X,D,MSE_epoch,weight_epoch):
    for this in range(epoch):
        weight_epoch[this,:]=W_old.reshape(dim)
        W_new = Weight_update(lr,W_old,X,D)
        W_old = W_new
        MSE_this =get_Loss(D,W_new,X)
        MSE_epoch[this,:]=MSE_this
```

```python
gra_de(W_old,lr,X,Y,Loss_epoch,weight_epoch)
log_Loss= np.log(Loss_epoch)
```

```python
def get_error(W,X,D):
    return D-np.dot(X,W)
def Weight_update2(lr,W_old,X,D,l_0):

    error=get_error(W_old,X,D)
    W_new = W_old+lr*2*np.dot(X.T,error)
    n0 = np.sum(W_new!=0)


    if n0 < l_0:
        return W_new
    else:
        drop = n0-l_0
        arr_in= np.absolute(W_new)

        ind =np.argsort(arr_in,axis=0)  

        index_drop = ind[0:drop]
        for item in index_drop:
            W_new[item]=0

        return W_new
def get_Loss(D,W,X):
    y = np.dot(X,W)
    square=np.power(D-y, 2)

    return np.sum(square)#+np.sum(W!=0)

```

```python
epoch = 200
lr = 0.02
MSE_epoch =np.zeros([epoch ,1])
print(MSE_epoch.shape)
weight_epoch = np.zeros([epoch,dim])
print(weight_epoch.shape)
W_old=np.array([0,0,0,0,0]).reshape(dim,1)
```

```python
def PGD(W_old,lr,X,D,MSE_epoch,weight_epoch):
    for this in range(epoch):
        weight_epoch[this,:]=W_old.reshape(dim)
        W_new = Weight_update2(lr,W_old,X,D,2)
        W_old = W_new
        MSE_this =get_Loss(D,W_new,X)
        MSE_epoch[this,:]=MSE_this
```

```python
def get_error(W,X,D):
    return D-np.dot(X,W)
def Weight_update3(lr,W_old,X,D,l_1):

    error=get_error(W_old,X,D)
    dr = np.sign(W_old)
    W_new = W_old+lr*(2*(np.dot(X.T,error))- l_1*dr)
    return W_new

def get_Loss2(D,W,X,item):
    y = np.dot(X,W)
    square=np.power(D-y, 2)
    return np.sum(square)#+item*np.linalg.norm(W,ord=1)
def PGD(W_old,lr,X,D,MSE_epoch,weight_epoch,item):
    for this in range(epoch):
        weight_epoch[this,:]=W_old.reshape(dim)
        W_new = Weight_update3(lr,W_old,X,D,item)
        W_old = W_new
        MSE_this =get_Loss2(D,W_new,X,item)
        MSE_epoch[this,:]=MSE_this
```

```python
def Loss_all(W,X,D,lamda):
    y = np.dot(X,W) 
    square=np.power(D-y, 2)
    return np.sum(square) #+ lamda*np.linalg.norm(W,ord=1)

def soft_threshold(z, soft):
    W=z
    for i in range(z.shape[0]):
        item = float(z[i,:])
        if item > soft:
            W[i,:] = item- soft
        elif item < -soft:
            W[i,:] = item+ soft
        else:
            W[i,:] = 0
    return W

def soft_thresh(z, l):
    return np.sign(z) * np.maximum(np.abs(z) - l, 0)


def update_W(X, D, W,lamda,lr):
    z= W + 2*lr*np.dot(X.T,D-np.dot(X,W))

    W_new = soft_thresh(z, lamda*0.02)
    return W_new

def ProGD(W_old,lr,X,D,MSE_epoch,weight_epoch,lamda):
    for this in range(epoch):
        weight_epoch[this,:]=W_old.reshape(dim)
        W_new =update_W(X,D,W_old,lamda,lr)
        W_old = W_new
        MSE_this =Loss_all(W_new,X,D,lamda)
        MSE_epoch[this,:]=MSE_this
```
